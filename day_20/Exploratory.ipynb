```python
# Lets find the missing values first

data.isna()
```

```python
data.isnull().sum()
```

There are more null values in the column that we cannot see. Let's try another way to see.

```python
null_features = [features for features in data.columns if data[features].isnull().sum()>1]

null_features
```

```python
# Letssee how much missing values above column has

for features in null_features:
    print(f'{features}: {np.round(data[features].isnull().sum())}')
```

### Since there are lots of missing values. We need to find relationship between missing values and dependent features (sales_price) in our case.

```python
for feature in null_features:
    df = data.copy() # copied becuse it doesen't effect our data 

    # If the feature is null we represent it with 1 else with 0.
    df[feature] = np.where(df[feature].isnull(), 1, 0)

    df.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Missing values are clearly affecting our sale_price. So we need to fill those missing values with proper values, We will do that in feature engineering. 

## Numerical variables

```python
numerical_features = [feature for feature in data.columns if data[feature].dtypes != 'O']

print(f'Length of the numerical features is: {len(numerical_features)}')

data[numerical_features].head()
```

So, out of 81 features we 38 features are numerical type

### Temporal Variables(Eg: Datetime Variables)


From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering which is the next video.



```python
# List of variables that contain year features
year_feature = [year for year in numerical_features if 'Yr' in year or 'Year' in year]
year_feature
```

```python
# print unique values in year_feature

for year in year_feature:
    print(f'Unique values for {year} is {data[year].unique()}')
```

```python
# Let's see if there is relationship between year sold and the price

data.groupby('YrSold')['SalePrice'].median().plot()

plt.xlabel('Year sold')
plt.ylabel('median house price')
plt.title('Year sold vs Price')
```

```python
# Compare all year feature with sold_price

for year in year_feature:

    if year != 'YrSold':
        df = data.copy()

        # Difference between year variable and year the house is sold.

        df[year]=df['YrSold'] - df[year]


        plt.scatter(df[year], df['SalePrice'])
        plt.xlabel('feature')
        plt.ylabel('price')
        plt.show()
```

```python
# Numerical variables are basically two types: 1. continuous and discrete.

discrete = [dis_feature for dis_feature in numerical_features if len(data[dis_feature].unique())<25 and dis_feature not in year_feature + ['Id']]
discrete
```

```python
data[discrete].head()
```

```python
# Relationship between discrete features and price

for dis_feature in discrete:
    df = data.copy()

    df.groupby(dis_feature)['SalePrice'].median().plot.bar()

    plt.xlabel(dis_feature)
    plt.ylabel('Price')
    plt.title(dis_feature)

    plt.show()
```

```python
continuous = [cont_features for cont_features in numerical_features if cont_features not in discrete+year_feature+['Id']]
continuous
```

```python
data[continuous].head()
```

```python
# Lets analyze the continuous values by creating the histograms to understand the distribution. 

for cont_feature in continuous:
    df = data.copy()

    df[cont_feature].hist(bins=25)

    plt.xlabel(cont_feature)
    plt.ylabel('Count')
    plt.title(cont_feature)

    plt.show()
```
