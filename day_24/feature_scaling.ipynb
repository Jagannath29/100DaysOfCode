## Feature Selection
- The main aim of feature selection is to predict the house price based on various features which we will discuss as we go ahead



```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Libraries for feature selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# To visualize all the column in the dataset
pd.pandas.set_option('display.max_columns', None)
```

```python
# Lets load the training dataset that we preprocessed eariler and saved it as train.csv in day_23.

path = '/home/ubuntu/100DaysOfCoding/day_23/train.csv'

dataframe = pd.read_csv(path)
dataframe.head()
```

```python
# keeping dependent feature inside y_test
y_train = dataframe['SalePrice']
```

```python
# Drop id, sale_price feature from the dataframe

x_train = dataframe.drop(['Id', 'SalePrice'], axis=1)
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_selection_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) ## remember to set the seed, the random state in this function
feature_selection_model.fit(x_train, y_train)
```

```python
feature_selection_model.get_support()
```

```python
# let's print the total number of selected features

# This is how we can make list of selected features
selected_features = x_train.columns[(feature_selection_model.get_support())]

# let's print some status
print(f'Total features: {x_train.shape[1]}')
print(f'Selected features: {len(selected_features)}')

```

```python
x_train = x_train[selected_features]
```
