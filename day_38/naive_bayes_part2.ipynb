### Today I am going to build email spam classiier using naive bayes algorithm.  
- The dataset contains two columns message and category.
- Messages are normal mails, Category are spam and ham
    - If the output is spam, email is dangerous.
    - If the output is ham, email in not dangerous.

### I will use sklearn CountVectorizer to convert email text into a matrix of numbers and then use sklearn MultinomialNB classifier to train our model.


The model score with this approach comes out to be very high (around 99%). Sklearn pipeline allows us to handle pre processing transformations easily with its convenient api. 

```python
# Libraries

import pandas as pd
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

Ml understands only numbers so we need to convert the column into numbers somehow.

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df.head()
```

```python
# drop category column
df.drop('Category', axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

x_train_count = cv.fit_transform(x_train.values)
x_train_count.toarray()[:3] # it is message in our dataset converted into numbers
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
emails = [
        'Hey mohan, can we get together to watch footbal game tomorrow?', 
        'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
    ]
emails_count = cv.transform(emails)
model.predict(emails_count)
```

First message looks like positive, model pridict it as 0.

second message looks like negative, model pridict it as 1.




```python
x_test_vector = cv.transform(x_test)
model.score(x_test_vector, y_test)
```
