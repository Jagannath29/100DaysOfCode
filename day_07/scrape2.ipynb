```python
# importing libraries
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# url of programming language
html = urlopen("https://en.wikipedia.org/wiki/Comparison_of_programming_languages")
html
```

```python
# Converting the url to html
soup = BeautifulSoup(html, "html.parser")

soup
```

```python
# Finding table and rows in the url
table = soup.findAll("table", {"class":"wikitable"})[0]
rows = table.findAll("tr")
print(table)
```

```python
# program to read and write in "language.csv" file
with open("language.csv", "wt+", newline="") as f:
    writer = csv.writer(f)
    for i in rows:
        row = []
        for cell in i.findAll(["td", "th"]):
            row.append(cell.get_text())
        writer.writerow(row)
```
