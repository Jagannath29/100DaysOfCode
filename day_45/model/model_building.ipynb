```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

#### This CSV file is already preprocessed and saved

## One hot coding

```python
dummy_location = pd.get_dummies(df1['location'])
dummy_location.head()
```

```python
dummy_society = pd.get_dummies(df1['society'])
dummy_society.head()
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding = LabelEncoder()
encoding.fit(df1['area_type'])
```

```python
df1['label_area_type'] = encoding.transform(df1['area_type']) 
```

```python
df1.head()
```

Let's drop  area type column

```python
df1.drop('area_type', axis=1, inplace=True)
```

```python
df2 = pd.concat([dummy_location.drop('others', axis=1), dummy_society, df1], axis=1)
df2.head()
```

In Onehotcoding, if we have 1 less column from the total column it is fine because in one hot coding every rows is 0 except for 1 column. so, when we have 0 at every rows we can assume that the dropped column has value 1.

Since I have each and every unique location and society, it is safe to drop both columns.

```python
df2.drop(['society', 'location'], axis=1, inplace=True)
```

### Split the dataset for training and testing 

```python
x = df2.drop(['price'], axis=1) # dropping the target varaibl
y = df2['price']
```

```python
# libraries
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=
                                                    0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_train.head()
```

```python
x_test.shape
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test)[:5]
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_train,y_train)
```

Our training and testing score is nearly same. It means our model is overfitting. 

Let's use l1 and l2 regularization to reduce.

```python
from sklearn.linear_model import Lasso

las_reg = Lasso()
las_reg.fit(x_train, y_train)
```

```python
las_reg.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge

ri_reg = Ridge()
ri_reg.fit(x_train, y_train)
```

```python
ri_reg.score(x_test, y_test)
```

Let's apply other algorithm and find which is performing better

```python
from sklearn.model_selection import ShuffleSplit, RandomizedSearchCV

```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def best_algo(x,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

best_algo(x,y)
```

```python
best_algo(x,y)
```
