```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')

df.head()
```

```python
plt.scatter(df['age'], df['bought_insurance'], marker='+')
plt.xlabel('Age')
plt.ylabel('Insurance Bought')
```

The plot is giving just 0 and 1 because it is what inside our dataframe.

High number of age has bought insurance than the lower age.

## Let's have a quick view how logistic regression works.

```python
# Libraries
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# lets assign age and insurance bought to x,y
x = df[['age']]
y = df.bought_insurance
```

```python
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train
```

```python
x_test
```

### This is our 80% of the training dataset of just independent variable age only

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# fit the model
model.fit(X_train, y_train)
```

```python
x_test
```

```python
model.predict(x_test)
```

We have predicted corresponding y label from the above x_test.

As we saw in the above scatter plot, person with higher age is likely to buy insurance anc vice versa.

```python
# Lets see the score, it is nothing but the accuracy of the model

model.score(x_test, y_test)
```

Accuracy above 70% meaning our model is performing well. 

```python
# lets predict the probability if the person buys insurance or not.
model.predict_proba(x_test)
```

First column indicates low chance of buying and the other column indicates high chance of buying insurance.
