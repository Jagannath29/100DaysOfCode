# Popularity Based Recommender System

`Popularity based recommender system` : These are like trending videos in youtube. It stores bunch of popular things in one platform and is available for user. Some user may not like it.

```python
# let's see books with highest rating
rating_with_name = ratings.merge(book, on='ISBN')
rating_with_name.head()
```

```python
num_ratingin_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index() # reset_index returns the dataframe
```

```python
num_ratingin_df.rename(columns={"Book-Rating": 'num_rating'}, inplace=True)
```

```python
num_ratingin_df
```

```python
# Ensure 'Book-Rating' is numeric
rating_with_name['Book-Rating'] = pd.to_numeric(rating_with_name['Book-Rating'], errors='coerce')
```

```python
# Group by 'Book-Title' and calculate the mean rating
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
```

```python
# Renaming Book-rating into avg_rating
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_ratingin_df.merge(avg_rating_df, on='Book-Title')
popular_df

```

```python
# Books with num_rating > 250

popular_df = popular_df[popular_df['num_rating'] >= 250].sort_values('avg_rating', ascending=False).head(20)
popular_df
```

```python
# drop duplicate book-title
popular_df = popular_df.merge(book, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df
```

```python
# Take only useful column from popular_df
popular_df = popular_df[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_rating', 'avg_rating']]
```

```python
popular_df.head()
```

# Collaborative Filtering Based Recommender System


```python
# peoples and the number of books they rated

num_book_rated = rating_with_name.groupby('User-ID').count()['Book-Rating']
num_book_rated
```

user id 2 has rated 1 book, 8 has rated 17 books and so on.

```python
# people only with more than 200 book rated
book_rated_200 = num_book_rated > 200
```

```python
litterate_users = book_rated_200[book_rated_200].index
litterate_users
```

User_id of the users

```python
# Litterate users
filtered_users = rating_with_name[rating_with_name['User-ID'].isin(litterate_users)]
filtered_users
```

Out of 1149780, only 474007 are litterate users meaning user with rated more than 200 books.

```python
# book with 50 + rating

y = filtered_users.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = y[y].index

```

```python
famous_book
```

```python
final_rating = filtered_users[filtered_users['Book-Title'].isin(famous_book)]
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
# Fill pivot table 
pt.fillna(0, inplace=True)
```

```python
pt
```

### Time to find out the distance from one vector to another vector, We need it to measures the similarity between two vectors of an inner product space.

```python
from sklearn.metrics.pairwise import cosine_similarity

```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

It is the equlian distance of 706 books with 706 books 

```python
similarity_score[0]
```

Similarity score with index 0 and other remaining other. It returns 1 in first index because index 0 is exactly same when comparing with itself.

```python
# Function to recommend book

def recommend(book_name):
    # fetch index
    index = np.where(pt.index==book_name)[0][0]
    # 5 books with simlar items
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    # let's print the books using loop
    for i in similar_items:
        print(pt.index[i[0]])

```

```python
recommend("River's End")
```
