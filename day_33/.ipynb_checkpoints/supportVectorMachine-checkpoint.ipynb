## Support Vector Machine
- A support vector machine (SVM) is a supervised machine learning algorithm that classifies data by finding an optimal line or hyperplane that maximizes the distance between each class in an N-dimensional space.

### Hyperplane
- A hyperplane is a decision boundary that separates data points into different classes in a high-dimensional space.
- In two-dimensional space, a hyperplane is simply a line that separates the data points into two classes. 
- In three-dimensional space, a hyperplane is a plane that separates the data points into two classes. 



### Margin

- A margin is the distance between the decision boundary (hyperplane) and the closest data points from each class.


### Support Vectors

- They are the data points that lie closest to the decision boundary (hyperplane) in a Support Vector Machine (SVM).


### What is Gamma?

- Gamma is used when we use the Gaussian RBF kernel.
- Gamma is a hyperparameter which we have to set before training model. 
- Gamma decides that how much curvature we want in a decision boundary.



Gamma high means more curvature.

Gamma low means less curvature.

### What is C ?

- C- It is a hypermeter in SVM to control error. What does that mean to control error or margin? Let’s understand with visualization.


- You can see if we have low C means low error and if we have large C means large error.

```python
# Libraries
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
# Directories that iris contain
dir(iris)
```

```python
# feature_name

iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df.head()
```

```python
# append target column in the dataframe
df['target'] = iris.target
df.head()
```

We have target value 0, 1 and 2

```python
# let's find what are they
for i, name in enumerate(iris.target_names):
    print(f'{i} for {name}')
```

```python
# length of the dataframe
len(df)
```

```python
# From 50 to 100 we have versicolor
df[df.target == 1].head()
```

```python
# And from 100 to 150 we have virginica
df[df.target==2].head()
```

So in total we have 150 rows, each target contains 50 values.

```python
# Creating a new column name
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head(20)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['target', 'flower_name'], axis=1) #Dropping target column
```

```python
y = df[['target']] # Keeping target colume inside y
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)
```
